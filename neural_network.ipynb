```python
import numpy as np

# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh
def tanh(x):
    return np.tanh(x)

# ReLU
def relu(x):
    return np.maximum(0, x)
```

```python
# initialization of weights
np.random.seed(42)
w1 = np.random.randint(1, 10)
w2 = np.random.randint(1, 10)

# Forward pass
def forward(x1, x2, w1, w2):
    return x1 * w1 + x2 * w2

ypred = forward(3, 2, w1, w2)
print("Initial prediction:", ypred)
```

```python
y_actual = 17
error = (y_actual - ypred) ** 2
print("Initial Error:", error)
```

```python
learning_rate = 0.01

for epoch in range(20):
    ypred = forward(3, 2, w1, w2)
    error = (y_actual - ypred) ** 2

    # Gradients
    grad_w1 = -2 * 3 * (y_actual - ypred)
    grad_w2 = -2 * 2 * (y_actual - ypred)

    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

    print(f"Epoch {epoch+1}: ypred={ypred:.4f}, Error={error:.4f}")
```

```python
import matplotlib.pyplot as plt

errors = []
preds = []

w1, w2 = np.random.randint(1, 10), np.random.randint(1, 10)

for epoch in range(20):
    ypred = forward(3, 2, w1, w2)
    error = (y_actual - ypred) ** 2

    grad_w1 = -2 * 3 * (y_actual - ypred)
    grad_w2 = -2 * 2 * (y_actual - ypred)

    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

    preds.append(ypred)
    errors.append(error)

# Plot -- error progression
plt.plot(errors, label="Error")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.legend()
plt.show()

# Plot -- ypred progression
plt.plot(preds, label="Predicted y")
plt.axhline(y_actual, color="r", linestyle="--", label="Actual y")
plt.xlabel("Epochs")
plt.ylabel("y")
plt.legend()
plt.show()
```
